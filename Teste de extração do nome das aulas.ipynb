Código criado para fazer uma varredura nas URL de videos do vimeo para ver o nome de cada um deles.

Depois, se for interessante, entrar na URL específica e baixar o vídeo (que muitas vezes são aulas)

A maioria das aulas da plataforma Conquer estão salvas no Vimeo (senão todas).

Exemplo de URL:

https://player.vimeo.com/video/691819847 

https://player.vimeo.com/video/692038915

https://player.vimeo.com/video/691952419

https://player.vimeo.com/video/690901979

https://player.vimeo.com/video/690962373

1. Entrar na URL e extrair o título do site
2. Passar o titulo do site para um dataframe
3. Criar um laço de repetição que incrementa a URL para isso

```python
# importando bibliotecas necessárias

from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
# Criando indice crescente
indice = 1

# Criando numero para complementar url
numero = 690000000

# Criando laço para repetir o processo

while numero < 700000000:
    
    # Criando URL

    url_base = "https://player.vimeo.com/video/"
    url_complementar = str(numero)
    url_completa = url_base + url_complementar

    # Consultando a url com o requests e parseando ela com o BeautifulSoup

    page = requests.get(url_completa)
    soup = BeautifulSoup(page.content, "html.parser")

    # Criando variavel de títulos para adicionar o titulo da URL dentro

    titulo = soup.find("title").get_text()
    
    # Incrementando indice 
    
    indice += 1

    # Criando variável "lista" para adicionar o título extraído anteriormente dentro

    lista = pd.Series(data = titulo, index = [indice])
    
    
    # Printando lista após todo o loop (quando ela estiver completa)
    
    print(lista)
    
    # Incrementando numero
    numero += 1
    

    
```
